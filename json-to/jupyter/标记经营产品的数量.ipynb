```python
import pandas as pd

# 假设你的DataFrame变量名为df
# 先按照股票名称分组，然后统计每个股票的经营产品数量
product_count = df.groupby('股票名称')['经营产品'].nunique().reset_index()
# display(product_count)
# 添加标记列，将经营产品数量大于1的标记为"多产品"，否则标记为"单一产品"
product_count['标记'] = product_count['经营产品'].apply(lambda x: '多产品' if x > 1 else '单一产品')
display(product_count)
# 打印结果
# product_count.to_excel(r"D:\code\jupyter\json-to\企业经营产品数量.xlsx")
```

```python
import pandas as pd
# 假设你的DataFrame变量名为df
# 先按照股票名称分组，然后统计每个股票的经营产品数量
product_count = df.groupby('股票名称')['经营产品'].nunique().reset_index()
display(product_count)
# 将经营产品数量合并到原始DataFrame
df_result = pd.merge(df, product_count, on='股票名称', how='left')

# 添加标记列，表示经营产品数量
df_result['经营产品数量标记'] = df_result['经营产品_y'].apply(lambda x: f'{x}')

# 删除多余的列
df_result = df_result.drop(['经营产品_y'], axis=1)
display(df_result)
# 打印结果
# df_result.to_excel(r"D:\code\jupyter\json-to\标记产品数量.xlsx")
```
