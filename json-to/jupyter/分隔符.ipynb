### 根据分隔符分割行

```python
## 通过分隔符将行分割为更多行

def rowToRow(temp):
    temp_1=temp['输入关键词'].str.split('、',expand=True)
    temp_1=temp_1.stack()
    temp_1 = temp_1.reset_index(level=1,drop=True)
    temp_1.name='输入关键词'
    df_new = temp.drop(['输入关键词'], axis=1).join(temp_1)
    display(df_new)
    return df_new
```
