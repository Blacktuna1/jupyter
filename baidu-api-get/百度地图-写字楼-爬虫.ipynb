### 网格划分

```python
def generate_grid(top_left, bottom_right, grid_size):
    
    top_lat, left_lon = top_left
    bottom_lat, right_lon = bottom_right
    lat_grid_size, lon_grid_size = grid_size

    grid = []
    current_lat = top_lat

    while current_lat > bottom_lat:
        current_lon = left_lon
        while current_lon < right_lon:
            grid.append((current_lat, current_lon))
            current_lon += lon_grid_size

        current_lat -= lat_grid_size

    return grid

# Example: Generate a grid for Tianhe District, Guangzhou
tianhe_top_left = (23.1697, 113.3272)  # Top-left corner
tianhe_bottom_right = (23.1161, 113.4178)  # Bottom-right corner
grid_size = (0.001, 0.001)  # Adjust grid size as needed

tianhe_grid = generate_grid(tianhe_top_left, tianhe_bottom_right, grid_size)

# Print the first few grid coordinates for illustration
print(tianhe_grid)
```

```python
from tqdm import tqdm
from concurrent.futures import ThreadPoolExecutor, as_completed

def process_grid(grid):
    data_one = loads(search_poi('大厦', '写字楼', '广州', deal_lan(grid))).get('results')
    data_two = loads(search_poi('广场', '购物中心,百货商场,休闲广场', '广州', deal_lan(grid))).get('results')
    data_three = loads(search_poi('中心', '写字楼', '广州', deal_lan(grid))).get('results')

    rows = []

    if data_one and data_two and data_three:
        for poi_one, poi_two, poi_three in zip(data_one, data_two, data_three):
            rowOne = {'name': poi_one.get('name'), 'address': poi_one.get('address'),
                      'loc': poi_one.get('location'), 'uid': poi_one.get('uid'),
                      'tag': poi_one.get('detail_info').get('tag')}
            rowTwo = {'name': poi_two.get('name'), 'address': poi_two.get('address'),
                      'loc': poi_two.get('location'), 'uid': poi_two.get('uid'),
                      'tag': poi_two.get('detail_info').get('tag')}
            rowThree = {'name': poi_three.get('name'), 'address': poi_three.get('address'),
                        'loc': poi_three.get('location'), 'uid': poi_three.get('uid'),
                        'tag': poi_three.get('detail_info').get('tag')}
            rows.extend([rowOne, rowTwo, rowThree])
    return rows

# tianhe_grid 经纬度网格
data_list = []

# 初始化进度条
pbar = tqdm(total=len(tianhe_grid), desc="Processing")

# 设置线程池大小，根据需要调整
max_workers = 4

# 使用 ThreadPoolExecutor 实现多线程
with ThreadPoolExecutor(max_workers=max_workers) as executor:
    futures = [executor.submit(process_grid, grid) for grid in tianhe_grid]

    # as_completed 返回一个生成器，在任务完成时生成结果
    for future in as_completed(futures):
        try:
            rows = future.result()
            data_list.extend(rows)
        except Exception as e:
            print(f"An exception occurred: {e}")

        # 更新进度条
        pbar.update(1)

# 关闭进度条
pbar.close()
```
