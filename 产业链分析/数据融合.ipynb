```python
dataframeA_upper = df_2[['股票名称', '股票全称', '社会统一信用代码', '经营产品', '股票代码', '上游关键词']].copy()
dataframeA_upper.rename(columns={'上游关键词': '关键词'}, inplace=True)
dataframeA_upper['上下游关系'] = '上游'

# 创建下游数据框
dataframeA_lower = df_2[['股票名称', '股票全称', '社会统一信用代码', '经营产品', '股票代码', '下游关键词']].copy()
dataframeA_lower.rename(columns={'下游关键词': '关键词'}, inplace=True)
dataframeA_lower['上下游关系'] = '下游'

# 合并上游和下游数据框
df_A = pd.concat([dataframeA_upper, dataframeA_lower], ignore_index=True)

# 打印结果
display(df_A)
```

```python
df_A.to_excel(r"C:\Users\TUNGEE\Desktop\拆分.xlsx")
```

```python
# 复制原始数据框
df_last_2 = df_A.copy()

# 使用str.split和explode方法拆分关键词，并将其分开成多行
df_last_2['关键词'] = df_last_2['关键词'].str.split('、')
df_last_2 = df_last_2.explode('关键词')
```

```python
df_last_2.rename(columns={'关键词': '上下游产品'}, inplace=True)
df_last_2.to_excel(r"C:\Users\TUNGEE\Desktop\补充数据—last.xlsx")
```

```python
merged_dataframe = pd.concat([df_1, df_last_2], ignore_index=True)
```
